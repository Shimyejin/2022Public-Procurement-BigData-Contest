# 올리브영 크롤링

```python
#실행순서 1 - driver 및 url설정
from selenium import webdriver
import warnings
warnings.simplefilter(action = 'ignore')
import time
import pandas as pd

chrome_options = webdriver.ChromeOptions()
driver_path = "chromedriver.exe"
driver = webdriver.Chrome(driver_path, options=chrome_options)
url = 'https://www.oliveyoung.co.kr/store/store/getStoreMain.do'
driver.implicitly_wait(10)
driver.get(url)
```

```python
#실행순서 2 - 각 정보를 담기위한 리스트 생성
point_list = []
lat_list = []
lng_list = []
```

## 올리브영 양천구 크롤링

```python
#실행순서 3 - 양천구 검색버튼 클릭
driver.find_element_by_css_selector("#searchAreaTab > a").click()  #지역검색 클릭
driver.find_element_by_css_selector("#mainAreaList > option:nth-child(10)").click()  #서울특별시 클릭
driver.find_element_by_css_selector("#subAreaList > option:nth-child(20)").click()  #20, 21, 24(양천구, 영등포구, 종로구)로 바꿔가며 세개의 list에 각각 append
driver.find_element_by_css_selector("#searchAreaButton").click()
```

```python
#실행순서 4 - 지점별 정보 가져오기
point = driver.find_elements_by_css_selector("#areaStoreList > li > input.strNm") #지점명
latitude = driver.find_elements_by_css_selector("#areaStoreList > li > input.lat") #위도
longitude = driver.find_elements_by_css_selector("#areaStoreList > li > input.lng") #경도

point_list.extend([pnt.get_attribute("value") for pnt in point])
lat_list.extend([lat.get_attribute("value") for lat in latitude])
lng_list.extend([lng.get_attribute("value") for lng in longitude])
```

# 주거지(빌라, 오피스텔, 아파트) API 크롤링

```python
import requests
import json
```

```python
#서울시에 있는 모든 행정동
dong = pd.read_csv("dong.csv", encoding = 'cp949')
dong
```

```python
#리스트로 변환
dong2 = list(dong['행정동'])
dong2
```

## 빌라

```python
pointName = []
corX = []
corY = []

for addr in dong2:
    print(addr)
    
    #동마다 경위도 찾아서 경위도 기준으로 인근 빌라 찾기
    address = addr
    url =  'https://dapi.kakao.com/v2/local/search/address.json?query='+address
    headers = {'Authorization':'KakaoAK dc4c0f58233fec90a4fdcd35a897cf44'}
    result = json.loads(str(requests.get(url, headers = headers).text))
    
    x = result['documents'][0]['address']['x']  #각 동의 경도
    y = result['documents'][0]['address']['y']  #각 동의 위도
    
    convs = {}
    
    
    #빌라 위치 찾기
    for i in range(1, 46):
        headers = {'Authorization':'KakaoAK dc4c0f58233fec90a4fdcd35a897cf44'}
        
        #파라미터 가져오기
        params = {
            'x' : float(x),
            'y' : float(y),
            'radius' : 10000,  # 단위: m
            'page':i,
            'size':15,
            'sort':'distance'
        }
        keywords = '빌라'
        
        #api접근
        url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(keywords)
        places = requests.get(url, headers = headers, params = params).json()['documents']
        
        if places  == []:
            break
        else:
            convs[i] = places
    place_name = []
    cox = []
    coy = []
    
    
    for i in range(1, len(convs)):
        for a in range(len(convs[i])):
            place_name.append(convs[i][a]['place_name'])
            cox.append(convs[i][a]['x']) #빌라 경도
            coy.append(convs[i][a]['y']) #빌라 위도
    
    pointName += list(place_name)
    corX += list(cox)
    corY += list(coy)

#빈 데이터 프레임에 column저장
place = pd.DataFrame()
place['pointName'] = pointName
place['lon'] = corX
place['lat'] = corY
```

```python
place['type'] = 'billa'
```

# 대규모점포데이터 전처리

## 대규모점포 좌표 단위 변환

```python
from pyproj import Proj, transform
```

```python
big_market = pd.read_csv('fulldata_08_25_01_P_대규모점포.csv',encoding='ANSI')
```

```python
big_market = big_market.drop(['Unnamed: 29'],axis=1)
big_market
```

```python
proj_1 = Proj(init='epsg:2097')
proj_2 = Proj(init='epsg:4326')

converted = transform(proj_1, proj_2, big_market['좌표정보(x)'].values, big_market['좌표정보(y)'].values)
big_market['lon'] = converted[0]
big_market['lat'] = converted[1]
```

## 소재지 전체주소에서 빈 값을 도로명 전체주소로 대체

```python
import numpy as np
```
