```python
from pyproj import Proj, transform
import pandas as pd
```

```python
data = pd.read_csv('fulldata_08_25_01_P_대규모점포.csv',encoding='ANSI')
```

```python
data = data.drop(['Unnamed: 29'],axis=1)
```

```python
data
```

```python
proj_1 = Proj(init='epsg:2097')
proj_2 = Proj(init='epsg:4326')

converted = transform(proj_1, proj_2, data['좌표정보(x)'].values, data['좌표정보(y)'].values)
data['lon'] = converted[0]
data['lat'] = converted[1]
```

```python
data.to_csv('대규모점포.csv', index=False)
```
